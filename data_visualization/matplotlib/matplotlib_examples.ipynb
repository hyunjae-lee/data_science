```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
x
```

```python
y = x **2
y
```

```python
# Functional Method
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x,y)

plt.subplot(1, 2, 2)
plt.plot(y,x)
```

```python
# Object-oriented Method

## First
fig = plt.figure()

## Second
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
```

```python

## Third
axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('Larger plot')
axes2.set_title('Smaller plot')

fig
```

```python
fig = plt.figure(figsize = (8 , 4), dpi = 60)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig.savefig('my_pic.png', dpi = 200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X square')
ax.plot(x,x**3, label='X cubed')
#      (legend)
ax.legend()

#   
ax.set_xlim( [0, 10] )
```
